<font size=5>회전 초밥집에서는 담긴 접시의 색에 따라 초밥의 가격이 다르게 계산된다.<br>다음을 참고하여 회전 초밥의 비용을 영수증 형식으로 출력하는 프로그램을 완성해 보자.<br>

<br><br><br>

##입력예시


```
빨간색 접시의 개수를 입력하세요: 2
노란색 접시의 개수를 입력하세요: 3
파란색 접시의 개수를 입력하세요: 1
초록색 접시의 개수를 입력하세요: 2
```


##출력예시
```
========= 회전 초밥 영수증 =======
 접시     단가     개수       합계
빨간색   1,000       2       2,000
노란색   2,000       3       6,000
파란색   3,000       1       3,000
초록색   4,000       2       8,000
총 가격                     19,000
==================================
```



입력 = 4개

변수 = 개

input(), 이스케이프 시퀀스, 계산(합계)

````python
r = int(input("빨간색 접시의 개수"))
r_price = 1000
r_price_total = r * r_price
print("========회전 초밥 영수증========")
print("접시\t단가\t개수\t합계")
print(f"빨간색{r_price}\t{r}\t{r_price_total}")
````
